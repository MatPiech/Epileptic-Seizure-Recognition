```python
import pandas as pd
import numpy as np
from sklearn import preprocessing, model_selection, neighbors
```

Wczytujemy dane epileptic.csv - są to wartości odczytów EEG. W pliku mamy 23.6 sekundowe odczyty dla 500 unikatowych osób. Jedna seria odczytów składa się z 4097 próbek, co podzieleniu daje nam 178 próbek na sekundę (liczba kolumn odczytów). 

```python
df = pd.read_csv("epileptic.csv")
df.replace('?', -99999, inplace=True)
df.drop(['id'], 1, inplace=True)
#print(df['y'].head())
```

Tworzymy listę do binarnej klasyfikacji, wiedząc, że atak epilepsji występuje jedynie w przypadku wartości y równej 1 i to jej przypisujemy wartość 1. Gdy y przyjmuje wartości 2,3,4,5 nie występuje atak, dlatego w tym przypadku do listy verdict dodajemy wrtość 0. Następnie listę danych zamieniamy na "ramkę danych" i przypisujemy ją do nowej kolumny o nazwie 'verdict', jednocześnie usuwając niepotrzebną nam kolumne 'y'.  

```python
verdict = []
for y in df['y']:
    if y == 1:
        verdict.append(y)
    else:
        verdict.append(0)
        
verdict = pd.DataFrame(verdict)
df['verdict'] = verdict 
df.drop(['y'], 1, inplace=True)
print(df.head())
```

```python
X = np.array(df.drop(['verdict'], 1))
y = np.array(df['verdict'])
```

Dzielimy dane na część treningową i testową w stosunku 80% do 20%

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2)
```

Tworzymy klasyfikator w oparciu o K najbliższych sąsiadów

```python
clf = neighbors.KNeighborsClassifier()
```

Dopasowujemy klasyfikator do danych treningowych i sprawdzamy jego dokładność

```python
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
print(accuracy)
```
